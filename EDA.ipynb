# Variable de interes #

'budget': El presupuesto de una película puede tener un impacto significativo en su éxito y calidad de producción.
'popularity': La popularidad de una película puede influir en su éxito comercial y en la atención que recibe del público.
'revenue': Los ingresos generados por una película son un indicador clave de su éxito financiero.
'runtime': La duración de una película puede influir en la experiencia del espectador y en la comercialización de la misma.
'vote_average': La calificación promedio de votos de una película puede ser un indicador de su calidad y aceptación general.
'release_year': El año de lanzamiento de una película puede tener un impacto en su contexto histórico, influencia cultural y relevancia en relación con otras películas del mismo período.
'original_language': El idioma original de una película puede ser relevante para el análisis de la industria cinematográfica a nivel global y para comprender las preferencias de audiencia en diferentes regiones y países.

```python

# Seleccionar las variables de interés
features = ['budget', 'popularity', 'revenue', 'runtime',
            'vote_average', 'release_year']
X = df[features]
numeric_df = df[features]

# Calcular la matriz de correlación
corr_matrix = X.corr()


# Generar el mapa de calor de la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()


# Visualizar la matriz de correlación como un gráfico de dispersión
sns.pairplot(numeric_df)
plt.title('Relación entre variables numéricas')
plt.show()
```

### Observaciones acerca de las variables de interes ###


'budget': El presupuesto de una película parece tener un impacto significativo (0,75) en los ingresos ("revenue")

'popularity': Puntaje de popularidad tambien parece influir o tener relacion (0,46) con los ingresos ("revenue") casi en la misma media (0,45) que con el presupuesto ("budget")

'revenue': Los ingresos generados por una película parecen estar relacionado en mayor medida con el prespuesto y en menor medida el puntaje de popularidad

'runtime': La duración de una película, en este caso, no parece tener demaciada correlacion con las demas variables mas allá del puntaje promedio de las reseñas ("vote_average")

'vote_average': La calificación promedio de votos de una película esta relacionada casi por igual, en baja medida, con el puntaje de popularidad ("Popularity")y el tiempo de duracion ("runtime")

'release_year': El año de lanzamiento de una película parece tener una correlacion casi nula con el resto de las variables.
